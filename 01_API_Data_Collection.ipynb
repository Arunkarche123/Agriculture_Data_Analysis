```python
import requests
import pandas as pd
import time

API_KEY = "579b464db66ec23bdd000001a910e05595ff4b6259891847cfb6a007"

RESOURCE_ID = "9ef84268-d588-465a-a308-a864a43d0070"

BASE_URL = f"https://api.data.gov.in/resource/{RESOURCE_ID}"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

all_records = []

offset = 0
limit = 1000

while True:

    params = {
        "api-key": API_KEY,
        "format": "json",
        "offset": offset,
        "limit": limit
    }

    try:
        response = requests.get(
            BASE_URL,
            params=params,
            headers=headers,
            timeout=60
        )

        print(response.url)
        print(response.status_code)

        response.raise_for_status()

        data = response.json()

        records = data.get("records", [])

        if not records:
            print("Finished.")
            break

        all_records.extend(records)

        print(f"Total downloaded: {len(all_records)}")

        offset += limit

        time.sleep(2)

    except Exception as e:
        print(e)
        break

df = pd.DataFrame(all_records)

df.to_csv("AGMARKNET.csv", index=False)

print(df.shape)
```
