```python
import pandas as pd

market_df = pd.read_csv("Dataset2_Market_Locations.csv")

market_df.head()
```

```python
import pandas as pd
import requests
import time

weather_data = []

total = len(market_df)

print(f"Total Markets : {total}")

for i, row in market_df.iterrows():

    if pd.isna(row["latitude"]) or pd.isna(row["longitude"]):
        continue

    url = (
        f"https://archive-api.open-meteo.com/v1/archive?"
        f"latitude={row['latitude']}"
        f"&longitude={row['longitude']}"
        f"&start_date=2026-07-13"
        f"&end_date=2026-07-13"
        f"&daily=temperature_2m_mean,"
        f"relative_humidity_2m_mean,"
        f"precipitation_sum,"
        f"wind_speed_10m_max,"
        f"weather_code"
        f"&timezone=auto"
    )

    try:

        response = requests.get(url, timeout=30)

        data = response.json()

        daily = data["daily"]

        weather_data.append({
            "market": row["market"],
            "latitude": row["latitude"],
            "longitude": row["longitude"],
            "temperature": daily["temperature_2m_mean"][0],
            "humidity": daily["relative_humidity_2m_mean"][0],
            "rainfall": daily["precipitation_sum"][0],
            "wind_speed": daily["wind_speed_10m_max"][0],
            "weather_code": daily["weather_code"][0]
        })

    except Exception as e:

        print(f"Error : {row['market']}")

    if (i + 1) % 20 == 0:

        pd.DataFrame(weather_data).to_csv(
            "checkpoint_weather.csv",
            index=False
        )

        print(f"{i+1}/{total} completed")

    time.sleep(0.5)

weather_df = pd.DataFrame(weather_data)

weather_df.to_csv(
    "Dataset3_Weather.csv",
    index=False
)

print("\nWeather Dataset Created Successfully!")

print(weather_df.shape)

display(weather_df.head())
```
